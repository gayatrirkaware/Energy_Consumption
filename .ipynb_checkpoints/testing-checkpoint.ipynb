```python
import pickle
import json
import pandas as pd
import numpy as np
```

```python
with open('Artifacts\linear_regression.pkl', 'rb') as f:
    model = pickle.load(f)
model
```

```python
with open('Artifacts\label_enc_data.json', 'r') as f:
    label_data = json.load(f)
label_data
```

```python
column_names = model.feature_names_in_
column_names
```

```python
Building_Type = "Commercial"
Square_Footage = 2457
Number_of_Occupants = 12
Appliances_Used = 5
Average_Temperature = 35.3
Day_of_Week = "Weekday"
```

```python
test_array = np.zeros((1,column_names.size))

test_array[0][0] = label_data['Building Type'][Building_Type]
test_array[0][1] = Square_Footage
test_array[0][2] = Number_of_Occupants
test_array[0][3] = Appliances_Used
test_array[0][4] = Average_Temperature
test_array[0][5] = label_data['Day of Week'][Day_of_Week]
test_array
```

```python
df_test = pd.DataFrame(test_array, columns= column_names)
df_test
```

```python
prediction = model.predict(df_test)[0]
print('Predicted Price is:',np.around(prediction,3))
```
